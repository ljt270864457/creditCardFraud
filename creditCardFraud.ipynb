## 特征处理

```python
fig, ax = plt.subplots(1, 2, figsize=(18,4))

amount_val = df['Amount'].values
time_val = df['Time'].values

sns.distplot(amount_val, ax=ax[0], color='r')
ax[0].set_title('Distribution of Transaction Amount', fontsize=14)
ax[0].set_xlim([min(amount_val), max(amount_val)])

sns.distplot(time_val, ax=ax[1], color='b')
ax[1].set_title('Distribution of Transaction Time', fontsize=14)
ax[1].set_xlim([min(time_val), max(time_val)])



plt.show()
```

- 由于特征分布不均匀，所以建议使用[RobustScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.RobustScaler.html?highlight=robustscaler)进行特征缩放

    计算方式 (x-x.median)/IQR

```python
from sklearn.preprocessing import RobustScaler
df['scaled_Amount'] = RobustScaler().fit_transform(
    df['Amount'].values.reshape(-1, 1))
df['scaled_Time'] = RobustScaler().fit_transform(
    df['Time'].values.reshape(-1, 1))
df.drop(['Amount', 'Time'], axis=1, inplace=True)
df.head()
```

```python
import math
fig, ax = plt.subplots(7, 4, figsize=(18, 4))
for index in range(28):
    column = f'V{index+1}'
    value = df[column].values
    i = math.floor((index)/4)
    j = index % 4
    sns.distplot(value, ax=ax[i][j])
    ax[i][j].set_title(column, fontsize=14)
    ax[i][j].set_xlim([min(value), max(value)])

plt.show()
```

- 按照李萌同学说的，对于V1这种特征，在极小值存在着一些离群点，这种处理方式通常有两种
    1. 直接删除1.5IQR之外的数据
    2. 如果大于1.5IQR之外的数据，可以当做1.5IQR来做，当然1.5这个值不是固定的，也可以是2，2.5等等
    
    先按照第二种方式进行处理，以1.5倍数来做

```python
def get_quantile_range(df, column, up_weight=1.5, down_weight=1.5):
    '''
    计算某一列的允许的最小值和最大值
    '''
    p1 = df[column].quantile(0.25)
    p3 = df[column].quantile(0.75)
    IQR = p3-p1
    min_value = p1-1.5*IQR
    max_value = p3+1.5*IQR
    return min_value, max_value


get_quantile_range(df, 'V1')
```

```python
columns = [f'V{i+1}' for i in range(28)]
for column in columns:
    '''
    替换上下限
    '''
    min_value, max_value = get_quantile_range(df, column)
    df.loc[df[column] < min_value, column] = min_value
    df.loc[df[column] > max_value, column] = max_value
```

```python
import math
fig, ax = plt.subplots(7, 4, figsize=(18, 4))
for index in range(28):
    column = f'V{index+1}'
    value = df[column].values
    i = math.floor((index)/4)
    j = index % 4
    sns.distplot(value, ax=ax[i][j])
    ax[i][j].set_title(column, fontsize=14)
    ax[i][j].set_xlim([min(value), max(value)])

plt.show()
```

```python
# 现在基本上各个特征服从正态分布，将其使用zscore进行特征缩放
from sklearn.preprocessing import StandardScaler
for column in columns:
    new_column = f'scaled_{column}'
    df[new_column] = StandardScaler().fit_transform(df[column].values.reshape(-1, 1))
df.head()
```

```python
columns_scaled = [f'scaled_V{i+1}' for i in range(28)]
columns_scaled.extend(['scaled_Time','scaled_Amount','Class'])
df_scaled = df[columns_scaled]
df_scaled.head()
```

## 划分数据集

```python
from sklearn.model_selection import train_test_split
# 使用分层抽样保证正负样本比例相同
train, test = train_test_split(
    df_scaled, test_size=0.2,  stratify=df_scaled['Class'])
```

## 解决样本不平衡
### 欠采样使数据正负样本数量保持一致

```python
print(train['Class'].value_counts())
print(test['Class'].value_counts())
```

```python
train_class_0 = train[train['Class']==0].sample(n=394)
train_class_1 = train[train['Class']==1]
train_sample = pd.concat([train_class_0,train_class_1])

X_train_sample = train_sample.drop('Class', axis=1).values
y_train_sample = train_sample['Class'].values.reshape(-1,1)

X_test = test.drop('Class', axis=1).values
y_test = test['Class'].values.reshape(-1,1)
sns.countplot(train_sample['Class'])
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

classifiers = {
    "LogisiticRegression": LogisticRegression(),
    "KNearest": KNeighborsClassifier(),
    "Support Vector Classifier": SVC(),
    "DecisionTreeClassifier": DecisionTreeClassifier()
}
```

```python
for key, classifier in classifiers.items():
    classifier.fit(X_train_sample, y_train_sample)
    training_score = cross_val_score(classifier, X_train_sample, y_train_sample, cv=5)
    print("Classifiers: ", classifier.__class__.__name__, "Has a training score of", round(training_score.mean(), 2) * 100, "% accuracy score")
```
